```python
from astroML.datasets import fetch_dr7_quasar
import matplotlib.pyplot as plt

# Fetch the quasar data
data = fetch_dr7_quasar()

# select the first 10000 points
data = data[:10000]

z = data['redshift']
plt.hist(z, bins=int(len(z)/50))
plt.show()

```

```python
import numpy as np

# Compute histogram
bin_content, bin_edges = np.histogram(z, bins=int(len(z)/50))

# Compute cumulative distribution function (CDF)
cdf_data = np.cumsum(bin_content)  # NumPy cumulative sum

# Use bin centers instead of bin edges for plotting
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

# Plot CDF
plt.plot(bin_centers, cdf_data, linestyle='-')
plt.xlabel("Redshift (z)")
plt.ylabel("Cumulative Count")
plt.title("Quasar Redshift CDF")
plt.grid()
plt.show()
```
