```python
import numpy as np
import matplotlib.pyplot as plt


# Immaginiamo che il nostro modello abbia un solo parametro theta
# I dati osservati (finti) sono generati da una normale centrata su 2

true_theta = 2.0
observed_data = np.random.normal(loc=true_theta, scale=1.0, size=50)

# Definiamo la likelihood: P (dati | theta)
def likelihood(theta):
    return np.prod(np.exp(-(observed_data - theta)**2 / 2))

# Definiamo la prior: P(theta)
# Semplice prior gaussiana centrata in 0
def prior(theta):
    return np.exp(-theta**2 / (2 * 5**2))

# La funzione non normalizzata della posteriori: P(theta | dati) ‚àù likelihood * prior
def posterior(theta):
    return likelihood(theta) * prior(theta)

# Valutiamo la posteriori in un range di theta
theta_vals = np.linspace(-1, 5, 500)
posterior_vals = np.array([posterior(t) for t in theta_vals])

# Normalizziamo solo per visualizzare
posterior_vals /= np.max(posterior_vals)

# Plottiamo la "vera" posteriori (che normalmente non conosceremmo graficamente)
plt.plot(theta_vals, posterior_vals, label="Posteriori (non normalizzata)")
plt.axvline(true_theta, color='red', linestyle='--', label="vero theta")
plt.xlabel("theta")
plt.ylabel("posteriori relativa")
plt.title("PDF a posteriori (non conosciuta in forma chiusa, ma valutabile)")
plt.legend()
plt.grid(True)
plt.show()

```
