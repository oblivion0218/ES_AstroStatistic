```python
import numpy as np
from matplotlib import pyplot as plt

p_RR = 0.5
p_SS = 0.9
p_SR = 0.5 #probability of sunny given rain
p_RS = 0.1 #probability of rain given sunny

Cloudy = 0  # number of days in witch was ...
Sunny = 0

prob_Sunny = [] # probabiliy of  ... for each day
Prob_Cloudy = []

N = 10000 #numer of days I want to simulate

W = 0 #weather=0  ==  rainy day , weather=1 == sunny day

for day in range(1 , N+1):

    p = np.random.uniform(0,1) # p is equal to probability
    if ( W == 0 ):
        
        if p <= p_RR:
            Cloudy += 1
            W = 0
        else :
            Sunny += 1
            W = 1

    else : # W == 1

        if p <= p_SS:
            Sunny += 1
            W = 1
        else :
            Cloudy += 1
            W = 0

    Prob_Cloudy.append(Cloudy/day)
    prob_Sunny.append(Sunny/day)  
            

Cloudy = Cloudy/N
Sunny = Sunny/N

print("Probability sunny = " , Sunny)
print("probability cloudy = " ,Cloudy)
```

```python
plt.plot(range(N), prob_Sunny, label = "Sunny")
plt.plot(range(N), Prob_Cloudy, label = "Rainy")
plt.xlabel("day")
plt.ylabel("value")
plt.legend()
plt.show()
```

```python
from astroML import stats as astroMLstats

def bins_num (X) :

    sigmaG2 = astroMLstats.sigmaG(X) # quantile at 75 - quantile at 25
    binsize = 2.7 * sigmaG2 / (len(X)**(1/3)) # Freedman-Diaconis
    N = (max(X) - min(X))/binsize

    return int(N+2)
```

```python
import astroML

plt.hist(prob_Sunny, bins_num(prob_Sunny), label = "sunny probability as histogram")
plt.xlabel("Probability")
plt.ylabel("values")
plt.legend()
plt.draw()

print("the median of the histogram is: ", np.median(prob_Sunny), " +/- ", astroML.stats.sigmaG(prob_Sunny))
```

```python
cut = 2000

prob_Sunny = prob_Sunny[cut:]
Prob_Cloudy = Prob_Cloudy[cut:]

plt.plot(range(N-cut), prob_Sunny, label = "Sunny")
plt.plot(range(N-cut), Prob_Cloudy, label = "Rainy")
plt.xlabel("day")
plt.ylabel("value")
plt.legend()
plt.show()
```
