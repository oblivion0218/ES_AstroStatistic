# Kicking horses
A famous early application of low-number statistics was an analysis of Prussian cavalryman horse-kick deaths by Bortkiewicz in 1898.

He studied the distribution of 122 men kicked to death by horses among 10 Prussian army corps within 20 years (so a total number of corpes of 200). He recorded the number of corps with a given number of deaths:
| Number of deaths | Number of groups |
|-----------------|----------------|
| 0               | 109            |
| 1               | 65             |
| 2               | 22             |
| 3               | 3              |
| 4               | 1              |

- Plot the resulting probability distribution. 
- Careful with the normalization
How does it look like? Again try to fit it by hand.

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import poisson
```

```python
N=200  #Number of observations
n_deaths=np.array([0,1,2,3,4])
n_groups=np.array([109,65,22,3,1])
prob=n_groups/N  
plt.scatter(n_deaths, prob, color='red', s=40, marker='o', label='Data')
plt.title("Probability Distribution")
plt.xlabel("Number of Deaths")
plt.ylabel("Probability")
plt.legend()
```

```python
plt.scatter(n_deaths, prob, color='red', s=40, marker='o', label='Data')
for mu in [0.3, 0.6, 0.9, 1.2]:
    model=poisson.pmf(n_deaths, mu)
    plt.plot(n_deaths, model, linestyle='-', label='$\mu='+str(mu)+"$")
plt.title("Probability Distribution")
plt.xlabel("Number of Deaths")
plt.ylabel("Probability")
plt.legend()
```

```python
mu_correct=np.sum(n_deaths*n_groups)/N
model=poisson.pmf(n_deaths, mu_correct)
plt.scatter(n_deaths, prob, color='red', s=40, marker='o', label='Data')
plt.plot(n_deaths, model, linestyle='-', color='blue', label="Poisson Fit")
plt.title("Probability Distribution")
plt.xlabel("Number of Deaths")
plt.ylabel("Probability")
plt.legend()
```
